```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

data = pd.read_csv(r"C:\Users\SHIN Soyeon\Desktop\DSL\volume_investors.csv", encoding='cp949')
data['Date'] = pd.to_datetime(data['Date'])
data['Month'] = [x.month for x in data['Date']]
data['Year'] = [x.year for x in data['Date']]
data['Day'] = [x.day for x in data['Date']]

data
```

```python
a= pd.DataFrame(data[data['Month']==12].groupby('Day')['Individual'].mean()).reset_index()
a
plt.bar(a['Day'], a['Individual'])
plt.grid(visible=True, axis = 'y')
plt.style.use('seaborn')
plt.show()
```

```python
a = pd.DataFrame(data[data['Month']==12].groupby('Day')['Institution'].mean()).reset_index()
a
plt.bar(a['Day'], a['Institution'])
plt.grid(visible=True, axis = 'y')
plt.style.use('seaborn')
plt.show()
```

```python
rec = data[(data["Date"]>="2015-12-01")&(data["Date"]<="2022-12-29")]
rec[rec['Month']==12].groupby('Day')['Individual'].mean().plot(kind='bar', color='lightcoral', legend=True)
rec[rec['Month']==12].groupby('Day')['Institution'].mean().plot(kind='bar', color='lightsteelblue')


plt.xlabel('Day')
plt.ylabel('Net Purchase')
plt.legend()
plt.show()
```
