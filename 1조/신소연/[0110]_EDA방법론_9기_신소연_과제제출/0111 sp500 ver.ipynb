```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sp500_df = pd.read_csv(r"C:\Users\SHIN Soyeon\Desktop\EDA Project\1조\Dataset\S&P500.csv", index_col=0)
sp500_df.rename(columns={'날짜':'Date','종가':'Close'}, inplace=True)
sp500_df
```

```python
from datetime import datetime
sp500_df["Date"] = pd.to_datetime(sp500_df["Date"])

#미국 45대 대선: 2016년 11월 8일

uspdt45 = sp500_df[(sp500_df["Date"]>="2016-10-08")&(sp500_df["Date"]<="2020-11-01")]
uspdt45['Close'] = [x.replace(",", "") for x in uspdt45['Close']]
uspdt45 = uspdt45.astype({'Close':'float64'})
plt.plot(uspdt45["Date"], uspdt45["Close"])
plt.xticks(rotation=45)
plt.show()
```
