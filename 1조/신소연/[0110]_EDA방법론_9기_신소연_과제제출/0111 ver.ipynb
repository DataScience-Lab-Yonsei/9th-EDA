```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
kospi_df = pd.read_csv(r"C:\Users\SHIN Soyeon\Desktop\EDA Project\1조\Dataset\kospi_data.csv")
kospi_df
```

```python
#20대 대선: 3월 9일

from datetime import datetime
kospi_df["Date"] = pd.to_datetime(kospi_df["Date"])

pdt20 = kospi_df[(kospi_df["Date"]>="2022-03-01")&(kospi_df["Date"]<="2022-08-31")]
plt.plot(pdt20["Date"], pdt20["Close"])
#plt.plot(pdt20["Date"]["2022-03-09"], "bo")
#특정 x값에 표시 - 아직
plt.xticks(rotation=45)
plt.show()
```

```python
#19대 대선: 2017년 5월 9일

pdt19 = kospi_df[(kospi_df["Date"]>="2017-05-01")&(kospi_df["Date"]<="2022-03-01")]
plt.plot(pdt19["Date"], pdt19["Close"])
plt.xticks(rotation=45)
plt.show()

#2020-02 에 큰 하락세
#특수한 상황(코로나)을 고려해야함
#전반기 약세 / 후반기 강세라는 것은 성립
```

```python
#18대 대선: 2012년 12월 19일

pdt18 = kospi_df[(kospi_df["Date"]>="2012-12-01")&(kospi_df["Date"]<="2017-05-01")]
plt.plot(pdt18["Date"], pdt18["Close"])
plt.xticks(rotation=45)
plt.show()

#전반기와 후반기가 거의 균등하다는 것을 알 수 있음
#후반기에 가장 상승하는 경향을 보이긴 함
#전반기<후반기는 성립한다고 보기 어려움
```

```python
#17대 대선: 2007년 12월 19일

pdt17 = kospi_df[(kospi_df["Date"]>="2007-12-01")&(kospi_df["Date"]<="2012-12-01")]
plt.plot(pdt17["Date"], pdt17["Close"])
plt.xticks(rotation=45)
plt.show()

#상반기 약세, 하반기 강세 성립
#하반기에 조금 더 안정적인 추세
```

```python
#16대 대선: 2002년 12월 19일

pdt16 = kospi_df[(kospi_df["Date"]>="2002-12-01")&(kospi_df["Date"]<="2007-12-01")]
plt.plot(pdt16["Date"], pdt16["Close"])
plt.xticks(rotation=45)
plt.show()

#상반기 약세, 하반기 강세 성립
#다만 꾸준히 성장하는 경향을 보임
#상반기에 하락했다가 하반기에 상승하는 추세가 아님
```
