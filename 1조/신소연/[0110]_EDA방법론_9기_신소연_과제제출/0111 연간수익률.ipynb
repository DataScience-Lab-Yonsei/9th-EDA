```python
#연간수익률
#(마지막일종가-시작일종가)/시작일종가
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
kospi_df = pd.read_csv(r"C:\Users\SHIN Soyeon\Desktop\EDA Project\1조\Dataset\kospi_data.csv")
kospi_df
```

```python
#19대 대선: 2017년 5월 9일

start_1st = kospi_df[kospi_df["Date"]=="2017-05-04"]
end_1st = kospi_df[kospi_df["Date"]=="2018-05-04"]
a = float(end_1st["Close"]) - float(start_1st["Close"]) / float(start_1st["Close"])

start_2nd = kospi_df[kospi_df["Date"]=="2018-05-04"]
end_2nd = kospi_df[kospi_df["Date"]=="2019-05-03"]
b = float(end_2nd["Close"]) - float(start_2nd["Close"]) / float(start_2nd["Close"])

start_3rd = kospi_df[kospi_df["Date"]=="2019-05-03"]
end_3rd = kospi_df[kospi_df["Date"]=="2020-05-04"]
c = float(end_3rd["Close"]) - float(start_3rd["Close"]) / float(start_3rd["Close"])

start_4th = kospi_df[kospi_df["Date"]=="2020-05-04"]
end_4th = kospi_df[kospi_df["Date"]=="2021-05-04"]
d = float(end_4th["Close"]) - float(start_4th["Close"]) / float(start_4th["Close"])

start_5th = kospi_df[kospi_df["Date"]=="2021-05-04"]
end_5th = kospi_df[kospi_df["Date"]=="2022-03-02"]
e = float(end_5th["Close"]) - float(start_5th["Close"]) / float(start_5th["Close"])

plt.plot([2017, 2018, 2019, 2020, 2021], [a, b, c, d, e])

#날짜별로 결측치(NaN이 아니라 그냥 값이 없음)로 되어 있는 것 같아서 이를 처리하는 과정이 필요한지 고민
#너무 노가다인가..싶어서 다른 방법이 궁금
```
