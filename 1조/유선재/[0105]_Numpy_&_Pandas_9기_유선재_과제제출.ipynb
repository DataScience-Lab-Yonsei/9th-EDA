```python
# Q10. 1*10의 array를 값이 랜덤이 되게 만들고 그 중에서 최댓값을 0으로 바꿔보세요
# 세션에서 명확하게 다루지 않은 내용입니다. 슬랙에 질문하거나 구글을 활용하시면 좋을 것 같습니다
```

```python
arr4 = np.random.random((1, 10)).flatten()
arr4
```

```python
arr4 = np.random.random((1, 10)).flatten()

maxIndex = 0

max = arr4[0]

for i in range(1, 10):
  if max <= arr4[i]:
    max = arr4[i]
    maxIndex = i

arr4[maxIndex] = 0

print(arr4)

```

```python
# Q18. 이제는 3개의 데이터프레임을 병합해보겠습니다
# 먼저 아래 3개의 데이터프레임을 생성하는 코드를 만들어보세요
# 주문 정보 ord_df
```

```python
from pandas import DataFrame
data1 = {'CUST_ID' : [100, 100, 200, 200, 300], 'ORD_NO':[11, 12, 13, 14, 15], 'PRDT_CD' : [9001, 9002, 9001, 9003, 9003]}
ord_df = DataFrame(data1)
ord_df
```

```python
# 고객 정보 cust_df
```

```python
data2 = {'CUST_ID':[100, 200, 300, 400], 'NAME':['kim', 'park', 'ahn', 'choi']}
cust_df = DataFrame(data2)
cust_df
```

```python
# 상품정보 prdt_df
```

```python
data3 = {'PRDT_CD':[9001, 9002, 9003], 'NM':['Notebook', 'TABLET', 'TV'], 'PRICE':[1000000,1500000,2000000 ]}
prdt_df = DataFrame(data3)
prdt_df
```

```python
# Q19. ord_df와 cust_df를 병합하여 아래와 같은 결과가 나오게 코드를 작성해보세요. ord_cust_df라는 변수로 저장하세요
```

```python
ord_cust_df = pd.merge(ord_df, cust_df)
ord_cust_df
```

```python
# Q20. 19번에서 만든 데이터프레임과 prdt_df를 병합하여 아래와 같은 결과가 나오게 코드를 작성해보세요. merge_df라는 변수로 저장하세요
```

```python
merge_df = pd.merge(ord_cust_df, prdt_df)
merge_df
```

```python
#Q21. 20번에서 만든 데이터셋으로 각 고객이 총 얼마를 지불하였는지 보여주는 코드를 작성해보세요
```
