```python
url = "https://comic.naver.com/index"

#driver 이용해서 주소 요청하기
driver.get(url)

# 단일추출 -> find_element, 다중추출 -> find_elements
#[전체 카테고리 태그 클릭]
webtoon = driver.find_element(By.XPATH, '//*[@id="menu"]/ul/li[2]/a/span').click()

#[세부 카테고리 태그 클릭]
sunday = driver.find_element(By.XPATH, '//*[@id="content"]/ul/li[8]/a').click()



#상품 이름 가져오기
items = driver.find_elements(By.CSS_SELECTOR, 'div.list_area.daily_img a' )
```
