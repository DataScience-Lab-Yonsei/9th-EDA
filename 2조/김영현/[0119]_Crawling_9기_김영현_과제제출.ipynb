```python
import numpy as np
import pandas as pd
```

```python
driver = webdriver.Chrome(chromedriver)

url = "http://www.paxnet.co.kr/stock/report/report?menuCode=2222"

driver.get(url)

time.sleep(3)

org_name = driver.find_elements(By.CSS_SELECTOR, 'strong.color-cate a')

data = []

for idx in range(len(org_name)):

    driver.find_element(By.XPATH, f'//*[@id="contents"]/div[1]/div[2]/ul/li[{idx+2}]/div[2]/p/a').click()
    name = driver.find_element(By.XPATH, '//*[@id="contents"]/div[1]/div[2]/div[1]/h1/a/span')
    a = driver.find_elements(By.CSS_SELECTOR, '#contents > div.cont-area > div.report-view > div.report-view-chart > div > div > div > div.pan-cont > div.data > ul > li span:nth-child(2)')

    lst1 = []
    lst1.append(name.text)
    for i in a:
        lst1.append(i.text)
    data.append(lst1)


    driver.back()
```

```python
data_set = pd.DataFrame(data,columns = ['org_name','yesterday', 'quantity','current','the transaction price','high','foreigner_rate','low','Market capitalization'])

data_set = data_set.set_index('org_name')
```
