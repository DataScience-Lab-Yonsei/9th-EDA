```python
# Q17.타이타닉 승객들을 Embarked와 성별을 기준으로 그룹화하고, 각 그룹 별로 평균 나이를 출력해보세요
```

```python
# Q18. 이제는 3개의 데이터프레임을 병합해보겠습니다
# 먼저 아래 3개의 데이터프레임을 생성하는 코드를 만들어보세요
# 주문 정보 ord_df
```

```python
ord_df=pd.DataFrame([[100,11,9001],[100,12,9002],[200,13,9001],[200,14,9003],[300,15,9003]],
                    index=[0,1,2,3,4], columns=['CUST_ID','ORD_NO','PRDT_CD'])
ord_df
```

```python
# 고객 정보 cust_df
```

```python
cust_df=pd.DataFrame([[100,'kim'],[200,'park'],[300,'ahn'],[400,'choi']], index=[0,1,2,3], columns=['CUST_ID', 'NAME'])
cust_df
```

```python
# 상품정보 prdt_df
```

```python
prdt_df=pd.DataFrame([[9001,'Notebook',1000000],[9002,'TABLET',1500000],[9003,'TV',2000000]],
                     index=[0,1,2], columns=['PRDT_CD', 'NM', 'PRICE'])
prdt_df
```

```python
# Q19. ord_df와 cust_df를 병합하여 아래와 같은 결과가 나오게 코드를 작성해보세요. ord_cust_df라는 변수로 저장하세요
```

```python
ord_cust_df=pd.merge(ord_df, cust_df, how='left', on='CUST_ID')
ord_cust_df
```

```python
# Q20. 18번에서 만든 데이터프레임과 prdt_df를 병합하여 아래와 같은 결과가 나오게 코드를 작성해보세요. merge_df라는 변수로 저장하세요
```

```python
merge_df=pd.merge(ord_cust_df, prdt_df, how='right', on='PRDT_CD')
merge_df
```

```python
#Q21. 19번에서 만든 데이터셋으로 각 고객이 총 얼마를 지불하였는지 보여주는 코드를 작성해보세요
```

```python
merge_df.groupby('NAME').sum().reset_index().set_index('NAME')[['PRICE']]
```
