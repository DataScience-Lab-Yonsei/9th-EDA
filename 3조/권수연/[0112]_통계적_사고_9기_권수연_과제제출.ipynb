```python
work_df = pd.read_csv("/gdrive/My Drive/상권배후지-직장인구.csv",encoding='cp949')
work_df.head()
```

```python
work_df1 = work_df[(work_df['기준_년_코드']==2022)&(work_df['기준_분기_코드']==3)]
```

```python
work_df1
```

```python
work_20 = work_df1['연령대_20_직장_인구_수']
work_30 = work_df1['연령대_30_직장_인구_수']
work_40 = work_df1['연령대_40_직장_인구_수']
work_20_sum = work_20.sum()
work_30_sum = work_30.sum()
work_40_sum = work_40.sum()
print('서울시 평균 20대 직장인구 수: ',work_20_sum/1090,'명')
print('서울시 평균 30대 직장인구 수: ',work_30_sum/1090,'명')
print('서울시 평균 40대 직장인구 수: ',work_40_sum/1090,'명')
```

30대 직장인구 수와 40대 직장인구 수가 비슷해보인다. 통계적으로 이들의 평균이 같은지 알아보자.

```python
work_30_list = work_30.tolist()
work_40_list = work_40.tolist()
```

```python
import scipy.stats

scipy.stats.ttest_ind(work_30_list, work_40_list, equal_var=False)
```

p=value의 값이 약 0.957 정도로 높은 값이 나왔다. 따라서 30대 직장인구 수와 40대 직장인구 수는 통계적으로 평균 차이가 없다.
