```python
# Q11. Titanic 파일을 불러와 titanic_df라는 변수에 저장하고(저장할 때 첫 번째 열은 index로 지정) 위에서 2개의 행만 출력해보세요

titanic_df = pd.read_csv('/content/drive/MyDrive/Titanic.csv', index_col=0)
titanic_df.head(2)
```

```python
# Q12. 먼저 결측치를 처리해보겠습니다. 결측치가 하나라도 있는 행은 제거하는 코드를 작성해보세요
# EDA 프로젝트 때는 결측치를 다른 값으로 채우거나 없애는 작업을 할 수 있지만 여기서는 없애고 진행하겠습니다

titanic_df.dropna(axis=0, inplace=False)
```

```python
# Q13. titanic_df에서 가장 어린 승객의 이름과 나이를 구할 수 있는 코드를 작성해보세요
# 답은 여러 개가 있을 수 있습니다. 

titanic_df_nona = titanic_df.dropna(axis=0, inplace=False)
titanic_df_nona.sort_values(by='Age', ascending=True, inplace=False)[['Name', 'Age']].head(1)
```

```python
# Q14. Fare 값이 가장 높은 값과 낮은 값의 차이를 구하는 코드를 작성해보세요
# 답은 여러 개가 있을 수 있습니다. 

print('가장 비싼 티켓과 가장 싼 티켓의 가격 차이: $', titanic_df_nona['Fare'].max() - titanic_df_nona['Fare'].min())
```

```python
# Q15. Fare 값 상위 5명의 평균 가격을 계산하는 코드를 작성해보세요

titanic_df_nona.sort_values(by='Fare', ascending=False, inplace=False)['Fare'].head(5).mean()
```

```python
# Q16. 타이타닉 승객 중 여성이면서 나이가 40세 이상인 고객들의 정보를 추출하는 코드를 작성해보세요
# 힌트 : 2개의 조건을 동시에 만족할 수 있게 인덱싱하려면 &를 활용하시는 것도 하나의 방법입니다

titanic_df[(titanic_df['Age'] >=40) & (titanic_df['Sex'] == 'female')]
```

```python
# Q17.타이타닉 승객들을 Embarked와 성별을 기준으로 그룹화하고, 각 그룹 별로 평균 나이를 출력해보세요

titanic_group = titanic_df.groupby(['Embarked', 'Sex'])
pd.DataFrame(titanic_group.mean()['Age'].reset_index())
```

```python
# Q18. 이제는 3개의 데이터프레임을 병합해보겠습니다
# 먼저 아래 3개의 데이터프레임을 생성하는 코드를 만들어보세요
# 주문 정보 ord_df

ord_df = pd.DataFrame([[100, 11, 9001], [100, 12, 9002], [200, 13, 9001], [200, 14, 9003], [300, 15, 9003]],
                      columns=['CUST_ID', 'ORD_NO', 'PRDT_CD'])
ord_df
```

```python
# 고객 정보 cust_df

cust_df = pd.DataFrame([[100, 'kim'], [200, 'park'], [300, 'ahn'], [400, 'choi']],
                      columns=['CUST_ID', 'NAME'])
cust_df
```

```python
# 상품정보 prdt_df

prdt_df = pd.DataFrame([[9001, 'Notebook', 1000000], [9002, 'TABLET', 1500000], [9003, 'TV', 2000000]],
                      columns=['PRDT_CD', 'NM', 'PRICE'])
prdt_df
```

```python
# Q19. ord_df와 cust_df를 병합하여 아래와 같은 결과가 나오게 코드를 작성해보세요. ord_cust_df라는 변수로 저장하세요

ord_cust_df = pd.merge(ord_df, cust_df,
                       how = 'left',
                       on = 'CUST_ID')
ord_cust_df
```

```python
# Q20. 18번에서 만든 데이터프레임과 prdt_df를 병합하여 아래와 같은 결과가 나오게 코드를 작성해보세요. merge_df라는 변수로 저장하세요

merge_df = pd.merge(ord_cust_df, prdt_df,
                    how = 'right',
                    on = 'PRDT_CD')
merge_df
```

```python
#Q21. 19번에서 만든 데이터셋으로 각 고객이 총 얼마를 지불하였는지 보여주는 코드를 작성해보세요

group_df = merge_df.groupby('NAME')
pd.DataFrame(group_df['PRICE'].sum())
```
