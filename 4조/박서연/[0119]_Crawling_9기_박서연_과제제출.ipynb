### 기상청 날씨알리미 앱 리뷰 크롤링

기상청이 '앱 홍보 콘테스트'까지 주최할 정도로 공을 많이 들인 날씨알리미 앱! 리뷰는 어떨까?

```python
# driver 이용해서 주소 요청하기

chromedriver = "C:/Users/izzle/Downloads/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(chromedriver)
url = 'https://play.google.com/store/apps/details?id=kr.go.kma.weatherapp'
driver.get(url)
driver.implicitly_wait(time_to_wait=2)

# 평가 및 리뷰 버튼 클릭
reviews = driver.find_element(By.XPATH, '//*[@id="yDmH0d"]/c-wiz[2]/div/div/div[1]/div[2]/div/div[1]/c-wiz[4]/section/header/div/div[2]/button/i').click()

#스크롤 높이 가져옴
last_page_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    #끝까지 스크롤 다운
    driver.execute_script("window.scrollTo(300, document.documentElement.scrollHeight);")
    time.sleep(2)
    # 새로 업데이트 된 스크롤 높이 가져옴
    new_page_height = driver.execute_script("return document.documentElement.scrollHeight")
    
    # 더 이상 새로고침이 되지 않는다면 break
    if new_page_height == last_page_height:
        break
    last_page_height = new_page_height
    
#동적 페이지 수집 도중에도 소스 코드를 정적 텍스트로 바꾸어 Beautiful Soup 이용 가능
html_source = driver.page_source

soup = BeautifulSoup(html_source, 'html.parser')

review_comments = soup.select('div.h3YV2d')

str_review_comments = []

for i in range(len(review_comments)):
    str_tmp = str(review_comments[i].text) 
    str_tmp = str_tmp.replace('\n', '')
    str_tmp = str_tmp.replace('\t', '')
    str_tmp = str_tmp.replace('               ', '')
   
    str_review_comments.append(str_tmp)
    
for i in range(len(str_review_comments)):
    print(str_review_comments[i])

driver.quit()

pd_reviewdata = {"Comment":str_review_comments}

review_pd = pd.DataFrame(pd_reviewdata)
```

### 기상청 날씨누리 단기예보 크롤링

오늘의 날씨를 크롤링해보자!

```python
chromedriver = "C:/Users/izzle/Downloads/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(chromedriver)
driver.get('https://www.weather.go.kr/w/weather/forecast/short-term.do')
time.sleep(5)
bsObject = BeautifulSoup(driver.page_source, 'html.parser')

date = datetime.datetime.now().strftime("%Y-%m-%d")

for item in bsObject.find_all('ul', {'data-date':date}):
    data = item.select('li')

    # 시각
    time = data[0].contents[1].string

    # 날씨
    weather = data[1].contents[1].string

    # 기온
    temperature = data[2].contents[1].contents[0].string

    # 체감온도
    sensible_temperature = data[3].contents[1].string

    # 강수량
    amount_of_precipitation = data[4].contents[1].string

    # 강수확률
    probability_of_precipitation = data[5].contents[1].string

    # 바람
    wind = data[6].contents[1].string + '' + data[6].contents[2].contents[0].string + data[6].contents[2].contents[1].string

    # 습도
    humidity = data[7].contents[1].string

    # 한파 영향
    cold_surge_effect = data[8].contents[1].string

    print(time, weather, temperature, sensible_temperature, amount_of_precipitation, probability_of_precipitation, wind, humidity, cold_surge_effect)
```

### 유튜브 댓글 추출

'연간 2만 3천 명 ‘초미세먼지’로 사망…정책 손질 시급' 뉴스 영상에 대한 사람들의 반응은 어떨까?

```python
import time

chromedriver = "C:/Users/izzle/Downloads/chromedriver_win32/chromedriver.exe" #다운받은 크롬드라이버 경로에 맞게 설정하기!
driver = webdriver.Chrome(chromedriver)

url = 'https://youtu.be/gh4Jbe_BUTg'
driver.get(url)
time.sleep(3)
driver.execute_script("window.scrollTo(0, 600)") 
time.sleep(1)

#스크롤 높이 가져옴
last_page_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    #끝까지 스크롤 다운
    driver.execute_script("window.scrollTo(300, document.documentElement.scrollHeight);")
    time.sleep(2)
    # 새로 업데이트 된 스크롤 높이 가져옴
    new_page_height = driver.execute_script("return document.documentElement.scrollHeight")
    
    # 더 이상 새로고침이 되지 않는다면 break
    if new_page_height == last_page_height:
        break
    last_page_height = new_page_height
    
#동적 페이지 수집 도중에도 소스 코드를 정적 텍스트로 바꾸어 Beautiful Soup 이용 가능
html_source = driver.page_source


soup = BeautifulSoup(html_source, 'html.parser')

youtube_comments = soup.select('yt-formatted-string#content-text')
youtube_likes = soup.select('#vote-count-middle')


str_youtube_comments = []
str_youtube_likes = []

for i in range(len(youtube_comments)):
    str_tmp = str(youtube_comments[i].text) 
    str_tmp = str_tmp.replace('\n', '')
    str_tmp = str_tmp.replace('\t', '')
    str_tmp = str_tmp.replace('               ', '')
   
    str_youtube_comments.append(str_tmp)
    
    str_l = str(youtube_likes[i].text.strip(' '))
    str_l = str_l.replace('\n', '')
    str_youtube_likes.append(str_l)
    
for i in range(len(str_youtube_comments)):
    print(str_youtube_comments[i], str_youtube_likes[i])

driver.quit()

pd_data = {"Comment":str_youtube_comments, "Likes":str_youtube_likes}

youtube_pd = pd.DataFrame(pd_data)
```
