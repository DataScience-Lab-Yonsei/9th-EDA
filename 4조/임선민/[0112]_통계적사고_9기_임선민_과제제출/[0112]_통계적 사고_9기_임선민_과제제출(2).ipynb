##t-test

```python
from scipy import stats
```

```python
# 서초구 4월 미세먼지 평균 농도
data1
```

```python
# 영등포구 4월 미세먼지 평균 농도
data2
```

```python
print(data_s)
print(data_y)
```

Two sample t-test
- 10년동안의 벚꽃이 가장 만개하는 4월 데이터를 사용함.
- 두 표본은 서초구의 10년간 4월 미세먼지 농도와 영등포구의 10년간 4월 미세먼지 농도이다.
- Ho: 두 표본의 평균이 같다.

```python
t_stat, p_value = stats.ttest_ind(data1,data2, equal_var=True, alternative="two-sided")
print("t-statistics : {}, p-value : {}".format(t_stat, p_value))
```

p-value < 0.05 이므로 귀무가설을 기각할 수 있고,

서초구와 영등포구의 4월 미세먼지 평균 농도는 다르다는 것을 결론 내릴 수 있다.

따라서, 상대적으로 4월 미세먼지 농도가 낮은 영등포구의 여의도 한강공원으로 피크닉 가는 것을 (서초구의 잠원 한강공원으로 피크닉을 가는 것보다) 추천한다.

##ANOVA

```python
#10년간 4월의 송파구 미세먼지 농도 데이터
data3 = data[(data['Place']=='송파구')]
data3 = data3[(data3['Month']-4)%100 == 0]
data3 = data3['Dust'][:10]
data3
```

```python
#10년간 4월의 광진구 미세먼지 농도 데이터
data4 = data[(data['Place']=='광진구')]
data4 = data4[(data4['Month']-4)%100 == 0]
data4 = data4['Dust'][:10]
data4
```

```python
#10년간 4월의 마포구 미세먼지 농도 데이터
data5 = data[(data['Place']=='마포구')]
data5 = data5[(data5['Month']-4)%100 == 0]
data5 = data5['Dust'][:10]
data5
```

```python
import statsmodels.api as scipy
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

정규성 검정

```python
#정규성 검정
print(stats.shapiro(data3))
print(stats.shapiro(data4))
print(stats.shapiro(data5))
```

정규성 검정: H0: 정규성 / H1: not 정규성
p-value > 0.05이므로 귀무가설을 기각할 수 없고, 정규성 가정이 성립한다.

```python
from scipy.stats import bartlett
print(stats.bartlett(data3,data4,data5))
print(stats.levene(data3,data4,data5))
```

등분산 검정: H0: 등분산 / H1: not 등분산
p-value > 0.05이므로 귀무가설을 기각할 수 없고, 등분산 가정이 성립한다.

One-way ANOVA

```python
F, p = stats.f_oneway(data3, data4, data5)
print( 'F-Ratio : {}'.format(F),
      'p-value : {}'.format(p),
      sep = '\n')
```

p-value < 0.05 이므로 귀무가설을 기각할 수 있고,
5개의 구 중 적어도 한 개의 구에서의 10년간 4월 미세먼지 평균 농도는 다른 구들의 데이터와 다르다는 결론을 내릴 수 있다.

이러한 ANOVA test는 데이터 분석할 특정 구를 선정하는 기준 또는 한강공원이 위치한 특정 구들을 분석할 때, 피크닉 장소 및 시간 추천할 때의 근거가 될 수 있다.
