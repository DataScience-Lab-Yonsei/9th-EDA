# 2. 데이터 전처리

## 1) 버스 노선별 평균 승하차 승객수 csv 생성

```python
#경로 -> 상대경로 사용
PATH = '버스노선별_승하차_인원/'

#모든 파일 이름을 리스트로 가져옴
file_list, csv_list = os.listdir(PATH), list()
for file in file_list :

    if file.split('.')[-1] == 'csv' :

        csv_list.append(file)

#boolean flag를 활용해 첫 데이터프레임을 생성, 이후엔 기존 데이터프레임과 새 데이터프레임 병합 방법 사용
boolean_flag_1 = True
boolean_flag_2 = True
for csv_file in csv_list :

#각각의 데이터 저장 시 인코딩 방식 혼용함 -> try, except 구문 활용하여 인코딩 방식이 다른 문제 해결
    try :

        df_whole = pd.read_csv(PATH + csv_file, encoding = 'cp949')

    except :

        df_whole = pd.read_csv(PATH + csv_file, encoding = 'utf-8-sig')

#표준 컬럼 리스트 설정
    if boolean_flag_1 :

        standard_column_list = list(df_whole.columns)
        boolean_flag_1 = False

#컬럼명, 컬럼 길이가 다른 부분 존재 -> 기존 컬럼 리스트를 복사해서 가져와 길이에 맞게 버스 코드 컬럼 인자를 삽입.
    print(standard_column_list)

    if list(df_whole.columns) != standard_column_list :

        try :

            print(1)
            df_whole = df_whole[standard_column_list]

        except :

            std_column_list_copy = standard_column_list.copy()
            std_column_list_copy.insert(1,'bus_code')
        
            df_whole.columns = std_column_list_copy
    else :

        pass

    df_whole = df_whole.groupby('노선번호').sum()[['승차총승객수','하차총승객수']]

#문자열 포맷팅을 이용해 연도, 월별 컬럼명 기입 자동화함
    df_whole['승하차_평균_{}'.format(csv_file.split('.')[-2][-4:])] = (df_whole['승차총승객수'] + df_whole['하차총승객수']) / 2
    df_whole = df_whole[['승하차_평균_{}'.format(csv_file.split('.')[-2][-4:])]].astype('int')
    df_whole = df_whole.reset_index()
    print(df_whole)

    if boolean_flag_2 :

        df_bus = df_whole
        boolean_flag_2 = False

    else :

        df_bus = pd.merge(df_bus, df_whole, on = '노선번호', how = 'outer')

list_name = list(df_bus['노선번호'])
list_bus =list()

#중복된 노선이지만 데이터 타입이 다른경우 -> try except 구문 이용하여 데이터 타입 변경
for name in list_name :

    if type(name) == str :

        name = name.split('(')[0].lstrip('0')

        try :

            name = int(name)

        except :

            pass

    list_bus.append(name)

df_bus['노선번호'] = list_bus
df_bus = df_bus.groupby('노선번호').sum().astype('int')

# 데이터 저장
df_bus.to_csv('df_bus.csv', encoding = 'utf-8-sig')
```

## 2) 지하철 역별 평균 승하차 승객수 csv 생성 및 부역명 제거

```python
PATH = '지하철호선별_역별_승하차_인원/'
file_list, csv_list = os.listdir(PATH), list()

for file in file_list :

    if file.split('.')[-1] == 'csv' :

        csv_list.append(file)

boolean_flag = True

for csv_file in csv_list :

    try :

        df_whole = pd.read_csv(PATH+csv_file, encoding = 'cp949')

    except :

        df_whole = pd.read_csv(PATH+csv_file, encoding = 'utf-8-sig')

    if str(df_whole.iloc[0,-1]) == 'nan' :

        column_list = df_whole.columns
        df_whole = df_whole.reset_index()
        df_whole.columns = ['사용일자', '노선명', '역명', '승차총승객수', '하차총승객수', '등록일자','NaN']
        df_whole = df_whole.dropna(axis = 1)

    df_whole = df_whole.groupby(['역명']).mean()[['승차총승객수','하차총승객수']]
    df_whole['승하차_평균_{}'.format(csv_file.split('.')[-2][-4:])] = (df_whole['승차총승객수'] + df_whole['하차총승객수']) / 2
    df_whole = df_whole[['승하차_평균_{}'.format(csv_file.split('.')[-2][-4:])]].astype('int')
    df_whole = df_whole.reset_index()

    if boolean_flag :

        df_subway = df_whole
        boolean_flag = False

    else :

        df_subway = pd.merge(df_subway, df_whole, on = '역명', how = 'outer')

#동일 역명 부역명 제거
list_name = list(df_subway['역명'])
list_subway = list()

for name in list_name:
    name = name.split('(')[0]
    list_subway.append(name)
    
df_subway['역명'] = list_subway
df_subway = df_subway.groupby('역명').sum()
df_subway = df_subway.astype('int')

# 데이터 저장
df_subway.to_csv('df_subway.csv', encoding = 'utf-8-sig')
```

```python
# function for set text color of positive
# values in Dataframes
def color_positive_red(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for positive
    strings, black otherwise.
    """
    if 0.025 <= val < 0.05:
        color = 'salmon'
    elif val < 0.025:
        color = 'darkred'
    else:
        color = 'black'
    return 'color: %s' % color
 
heatmap_13145.style.applymap(color_positive_red)
```
