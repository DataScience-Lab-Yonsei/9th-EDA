```python
import pandas as pd
import os
import numpy as np
from scipy import stats
```

```python
df_line = pd.read_csv('df_line.csv',encoding = 'utf-8-sig')
df_line['승하차_평균_2211'] = df_line['승하차_평균_2211_x']
df_line = df_line.drop(['승하차_평균_2211_x','승하차_평균_2211_y'], axis = 1)
```

```python
df_line
df_line.drop(['승하차_평균_2211_x','승하차_평균_2211_y'], axis = 1)
```

```python
df_line = df_line.set_index('노선명')
```

```python
df_line[df_line == 0]
```

# 9호선 2단계 개통(환승 노선: 2호선, 분당선)

```python
df_line_2_before_9 = df_line.loc['2호선', '승하차_평균_1501':'승하차_평균_1502']
df_line_2_after_9 = df_line.loc['2호선', '승하차_평균_1503':'승하차_평균_1602']
```

```python
print(df_line_2_before_9.mean(),df_line_2_after_9.mean())
df_line_2_before_9 = list(df_line_2_before_9)
df_line_2_after_9 = list(df_line_2_after_9)
```

```python
t_stat, p_value = stats.ttest_ind(df_line_2_before_9, df_line_2_after_9, alternative = 'two-sided')
```

```python
print('t_statistics : {}, p_value : {}'.format(t_stat, p_value))
```

### p-value > 0.05 --> 귀무가설 기각 불가능

```python
df_line_b_before_9 = df_line.loc['분당선', '승하차_평균_1501':'승하차_평균_1502']
df_line_b_after_9 = df_line.loc['분당선', '승하차_평균_1503':'승하차_평균_1602']
```
