```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
subway = pd.read_csv("C:/Users/심은조/Downloads/subway.csv", encoding='cp949')
```

```python
subway_oct2 = subway[(subway['수송일자'] == "2022-10-15") | (subway['수송일자'] == "2022-10-14") | (subway['수송일자'] == "2022-10-16") | (subway['수송일자'] == "2022-10-08") | (subway['수송일자'] == "2022-10-22")]
```

```python
subway_oct2['before'] = subway_oct2.iloc[:,7]+subway_oct2.iloc[:,8]+subway_oct2.iloc[:,9]+subway_oct2.iloc[:,10]+subway_oct2.iloc[:,11]+subway_oct2.iloc[:,12]+subway_oct2.iloc[:,13]+subway_oct2.iloc[:,14]+subway_oct2.iloc[:,15]+subway_oct2.iloc[:,16]
```

```python
subway_oct2['after'] = subway_oct2.iloc[:,17]+subway_oct2.iloc[:,18]+subway_oct2.iloc[:,19]+subway_oct2.iloc[:,20]+subway_oct2.iloc[:,21]+subway_oct2.iloc[:,22]+subway_oct2.iloc[:,23]+subway_oct2.iloc[:,24]+subway_oct2.iloc[:,25]+subway_oct2.iloc[:,26]
```

```python
sum_date2 = subway_oct2.groupby(['수송일자']) 
sum_date_before = sum_date2.sum()['before']
sum_date_after = sum_date2.sum()['after']
```

```python
#paired t-test
stat, p_val = stats.ttest_rel(sum_date_before[0], sum_date_after[0], alternative='greater')
print('statistic:', stat, '   p-value:', p_val)
```
