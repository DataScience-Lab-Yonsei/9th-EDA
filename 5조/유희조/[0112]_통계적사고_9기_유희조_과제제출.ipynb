# **남양주_출근시간대**

**10월**

```python
con_1 = namyangju_2210['시간'] == '05시~06시'
con_2 = namyangju_2210['시간'] == '06시~07시'
con_3 = namyangju_2210['시간'] == '07시~08시'
```

```python
am_namyangju_2210 = namyangju_2210.loc[con_1 | con_2 | con_3 ,  ]
am_namyangju_2210
```

**11월**

```python
con_4 = namyangju_2211['시간'] == '05시~06시'
con_5 = namyangju_2211['시간'] == '06시~07시'
con_6 = namyangju_2211['시간'] == '07시~08시'
```

```python
am_namyangju_2211 = namyangju_2211.loc[con_4 | con_5 | con_6 ,  ]
am_namyangju_2211
```

**10월, 11월 groupby**

```python
am_namyangju_2210_df = am_namyangju_2210.loc[:, ['년도', '월', '지역', '노선', '시간', '승차.2']]
am_namyangju_2211_df = am_namyangju_2211.loc[:, ['년도', '월', '지역', '노선', '시간', '승차.2']]
```

```python
am_namyangju_2210_df
```

```python
am_namyangju = pd.merge(am_namyangju_2210_df, am_namyangju_2211_df, how = 'left', on = '노선')
am_namyangju
```

```python
am_namyangju_group = am_namyangju.groupby(['시간_x', '노선'])
am_namyangju_group.mean()
```
