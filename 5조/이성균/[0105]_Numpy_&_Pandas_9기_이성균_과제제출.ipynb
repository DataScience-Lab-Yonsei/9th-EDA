#### Q18.

```python
ord_df = {'CUST_ID' : [100,100,200,200,300],
           'ORD_NO' : [11,12,13,14,15],
           'PRDT_CD' : [9001,9002,9001,9003,9003]}
ord_df = pd.DataFrame(ord_df)
ord_df
```

```python
cust_df = {'CUST_ID' : [100,200,300,400],
          'NAME' : ['kim','park','ahn','choi']}
cust_df = pd.DataFrame(cust_df)
cust_df
```

```python
prdt_df = {'PRDT_CD' : [9001,9002,9003],
          'NM' : ['Notebook','TABLET','TV'],
          'PRICE' : [1000000,1500000,2000000]}
prdt_df = pd.DataFrame(prdt_df)
prdt_df
```

#### Q19.

```python
ord_cust_df = pd.merge(ord_df, cust_df, how='left', on='CUST_ID')
ord_cust_df
```

#### Q20.

```python
merge_df = pd.merge(ord_cust_df,prdt_df)
merge_df
```

#### Q21.

```python
merge_df.groupby('NAME')[['PRICE']].sum()
```
