# **T-Test**

```python
from scipy import stats
```

타코벨 앞 정류장 - 서대문03 - 9월 vs. 11월

```python
t_stat, p_value = stats.ttest_ind(b['승차총승객수'], d['승차총승객수'], equal_var=False, alternative="two-sided")
print("t-statistics : {}, p-value : {}".format(t_stat, p_value))
```
