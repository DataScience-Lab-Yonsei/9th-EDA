### HTTP 응답 코드

내가 크롤링하고자 하는 사이트가 접속이 가능한지 불가능한지 알 수 있는 코드 <br>res.status_code가 다음 중 하나의 값을 반환<br>
1xx : 요청을 받았으며 작업이 진행 중인 경우 <br>
2xx : 크롤링이 가능한 상태 <br>
4xx(404) : 요청이 올바르지 않은 경우 <br>
5xx : 서버가 응답하지 않는 경우 <br><br>
즉 if문을 사용하여 res.status_code == 200 일 때에만 크롤링이 진행되도록 설정!

```python
import requests
url = 'https://www.gmarket.co.kr/e/bigs/newyear/tab1'
res = requests.get(url)
print('응답코드:', res.status_code)
```

```python
import requests
url = "https://nadocoding1.tistory.com"
res = requests.get(url)
print('응답코드:', res.status_code)
```

### G마켓 상품 크롤링 예시

<b>로봇으로 인식 방지</b><br>
로봇으로 인식되어 접속이 차단되는 경우를 방지하기 위한 표시 <br>
자신의 User-Agent는 <a href = 'https://www.whatismybrowser.com/detect/what-is-my-user-agent'>'what is my useragent'</a> 에서 확인

```python
#User-Agent 지정
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
```

```python
import requests
from bs4 import BeautifulSoup

#크롤링할 url 설정
url = 'https://corners.gmarket.co.kr/Bestsellers'

#url과 header 지정하여 정보 요청
res = requests.get(url, headers = headers)

#응답요청이 200이면 계속 실행, 아니면 중지
res.raise_for_status()

soup = BeautifulSoup(res.text, 'html.parser') 
```

```python
soup
```

```python
#select_one으로 단일추출
item = soup.select_one('div.best-list li a.itemname')
print(item.get_text())
```

```python
#select로 다중 추출 -> 리스트 형태로 반환하기 때문에 for문을 이용해 하나씩 뽑아내기
items = soup.select('div.best-list li a.itemname')
for item in items[:30]:
  print(item.get_text())
```

```python
#하위 페이지 링크도 같이 뽑아내기
items = soup.select('div.best-list li a.itemname')
for item in items[:30]:
  print(item.get_text(), item['href'])
```

## 유튜브 댓글 추출

```python
driver = webdriver.Chrome('chromedriver', options=options)
url = 'https://www.youtube.com/watch?v=iV6k4CNS3Hw'
driver.get(url)
time.sleep(3)
driver.execute_script("window.scrollTo(0, 600)") 
time.sleep(1)

#스크롤 높이 가져옴
last_page_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    #끝까지 스크롤 다운
    driver.execute_script("window.scrollTo(300, document.documentElement.scrollHeight);")
    time.sleep(2)
    # 새로 업데이트 된 스크롤 높이 가져옴
    new_page_height = driver.execute_script("return document.documentElement.scrollHeight")
    
    # 더 이상 새로고침이 되지 않는다면 break
    if new_page_height == last_page_height:
        break
    last_page_height = new_page_height
    
#동적 페이지 수집 도중에도 소스 코드를 정적 텍스트로 바꾸어 Beautiful Soup 이용 가능
html_source = driver.page_source


soup = BeautifulSoup(html_source, 'html.parser')

youtube_comments = soup.select('yt-formatted-string#content-text')
youtube_likes = soup.select('div#toolbar > span')


str_youtube_comments = []
str_youtube_likes = []

for i in range(len(youtube_comments)):
    str_tmp = str(youtube_comments[i].text) 
    str_tmp = str_tmp.replace('\n', '')
    str_tmp = str_tmp.replace('\t', '')
    str_tmp = str_tmp.replace('               ', '')
   
    str_youtube_comments.append(str_tmp)
    
    str_l = str(youtube_likes[i].text.strip(' '))
    str_l = str_l.replace('\n', '')
    str_youtube_likes.append(str_l)
    
for i in range(len(str_youtube_comments)):
    print(str_youtube_comments[i], str_youtube_likes[i])

driver.quit()

pd_data = {"Comment":str_youtube_comments, "Likes":str_youtube_likes}

youtube_pd = pd.DataFrame(pd_data)
```
